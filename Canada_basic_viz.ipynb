```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx'
df = pd.read_excel(path, sheet_name='Canada by Citizenship', skiprows = range(20), skipfooter = 2)
```

## Data clean

```python
# Clean up the dataset to remove columns that are not informative to us for visualization
df.drop(["AREA", "REG", "DEV", "Type", "Coverage"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# rename columns 
df.rename(columns={"OdName":"Country", "AreaName":"Continent", "RegName":"Region"}, inplace = True)
```

```python
df.head()
```

```python
# For consistency, ensure that all column labels of type string
all(isinstance(column, str) for column in df.columns)
```

```python
df.columns = list(map(str,df.columns))
df.columns
```

```python
all(isinstance(column, str) for column in df.columns)
```

```python
# set country name as index 
df.set_index("Country", inplace = True)
df.head()
```

```python
# add total column
df['Total'] = df.sum(axis=1)
df.head()
```

```python
# create a list of years from 1980 to 2013
years = list(map(str, range(1980,2014)))
print(years)
```

## Visualizing to inspect the data 

### Area plot

```python
# change plot style to ggplot-like
mpl.style.use('ggplot')
```

```python
#sort data in descending based on total column
df.sort_values(["Total"], ascending = False, axis = 0, inplace = True)
df_top5 = df.head()
df_top5
```

```python
# transpose the dataframe
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
# change the index values of df_top5 to type integer for plotting
df_top5.index = df_top5.index.map(int)
```

```python
# plot area 
ax = df_top5.plot(kind = "area", stacked = False, alpha = 0.35, figsize=(20,10))
ax.set_title("Immigration Trend of Top 5 Countries")
ax.set_xlabel("Year")
ax.set_ylabel("Number of immigrants")
plt.show
```

```python
# plot the area graphy for the 5 least countires of immigration to Canada
df.sort_values(["Total"], ascending = True, axis = 0, inplace = True)
df_least5 = df.head()
df_least5
```

```python
df_least5 = df_least5[years].transpose()
df_least5.head()
```

```python
df_least5.index = df_least5.index.map(int)
```

```python
ax = df_least5.plot(kind = "area", stacked = False, alpha = 0.45, figsize=(20,10))
ax.set_title("Immigration Trend of Least 5 Countries")
ax.set_ylabel("Number of Immigrants")
ax.set_xlabel("Years")
```

### Histograms

```python
# The frequency distribution of the number (population) of new immigrants to Canada in 2013
df["2013"].head()
```

```python
df["2013"].tail()
```

```python
count, bin_edges = np.histogram(df["2013"])
print(count)
print(bin_edges)
```

```python
ax=df["2013"].plot(kind="hist", figsize=(8,5), color = "green", xticks = bin_edges)
ax.set_title("Histogram of Immigration from 195 Countries in 2013")
ax.set_ylabel("Number of Countries")
ax.set_xlabel("Number of Immigrants")
```

```python
# import full colors to matplotlib 
import matplotlib
for name, hex in matplotlib.colors.cnames.items():
    print(name, hex)
```

```python
# check the immigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013
df_h2 = df.loc[["Denmark", "Norway", "Sweden"],years].transpose()
df_h2
```

```python
count, bin_edges = np.histogram(df_h2, bins = 15)
print(count)
print(bin_edges)
```

```python
# leave space for edges
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10
ax = df_h2.plot(kind="hist", bins = 15, xlim = (xmin, xmax), xticks=bin_edges, figsize=(10,6), color = ["steelblue", "olive", "lightcoral"])
ax.set_title("Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013")
ax.set_ylabel('Number of Years')
ax.set_xlabel('Number of Immigrants')
```

### Bar charts

```python
# check immigrants from iceland to Canada
df_iceland = df.loc["Iceland",years]
df_iceland.head()
```

```python
df_iceland.plot(kind="bar",figsize=(10,6), color = "lightblue")
plt.xlabel("Years")
plt.ylabel("Number of immigrants")
plt.title("Icelandic immigrants to Canada")

plt.annotate('',  # s: str. will leave it blank for no text
             xy=(32, 70),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

plt.annotate("2008-2011 Financial Crisis", 
             xy = (32,70),
             rotation = 72.5, 
             va= 'bottom', # want the text to be vertically 'bottom' aligned
             ha = 'left' ) # alligenment 

plt.show
```

```python
# Total number of immigrants to Canada from the top 15 countries
df.sort_values("Total", ascending = False, inplace = True)
df_top15 = df.head(15)
df_top15 = df_top15["Total"]
```

```python
df_top15.plot(kind="barh", figsize = (12, 12), color = "salmon")
plt.title("Top 15 Countries of Immigrants to Canada from 1980 to 2013")
plt.xlabel("Number of Immigrants")

for index, value in enumerate(df_top15):
    label = format(int(value), ',')
    # place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)  
    plt.annotate(label, xy = (value - 47000, index - 0.01), color = 'snow')
plt.show()
```
